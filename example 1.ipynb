```python
import pandas as pd

from pandaseda.eda import Describe, correlation_heat_map, correlation, distplots
```

# Example Based on "Titanic: Machine Learning from Disaster" dataset

```python
titanic = pd.read_csv('tables/titanic.csv')
desc = Describe(titanic)
```

```python
desc.display(sort='nans')
```

```python
distplots(titanic, ['Pclass', 'Fare', 'Age', 'SibSp'])
```

```python
desc.countplot(nuniques = 10, hue = 'Survived')
```

```python
correlation(titanic, 'Survived', thresh=0.1, method='spearman')
```

```python
correlation_heat_map(titanic, figsize = (20,20))
```

```python
desc.get_columns(number_of_nuniques = 2, mode = 'equal')
```
